```python
import pandas as pd

# Load your raw data
df = pd.read_csv('C:\\Users\\hp\\insurance-risk-analytics\\data\\insurance_data.csv')

# --- CLEANING LOGIC ---
# Example: Drop duplicates and fill missing numeric values
df_cleaned = df.drop_duplicates()
df_cleaned.fillna(df_cleaned.select_dtypes(include='number').mean(), inplace=True)

# Save the cleaned version
df_cleaned.to_csv('C:\\Users\\hp\\insurance-risk-analytics\\data\\insurance_data_cleaned.csv', index=False)
print("Cleaned data saved to C:\\Users\\hp\\insurance-risk-analytics\\data\\insurance_data_cleaned.csv")
```
